```python
import pandas as pd

data = pd.read_csv('../data/london_table.csv')
```

```python
station_id_set = list(set(data['station_id'].values))
target_list = ['PM2.5', 'PM10']
```

```python
import utils
import datetime

result = pd.DataFrame(columns=['station_id', 'utc_time', 'PM2.5', 'PM10'])

for station in station_id_set:
    temp_df = pd.DataFrame([[0 for i in range(4)] for i in range(365*24)], columns=['station_id', 'utc_time', 'PM2.5', 'PM10'])
    temp_df['station_id'] = [station for i in range(365*24)]
    
    for target in target_list:
        # check progress
        print(station, target)
        
        data_subset = data[data['station_id'] == station][['utc_time', target]]
        data_subset = data_subset[data_subset['utc_time'] < '2018-01-01']
        target_result = []
        X, y = utils.fillNA(list(data_subset['utc_time'].values), list(data_subset[target].values))
        
        temp_df[target] = y
    temp_df['utc_time'] = X
        
    result = pd.concat((result, temp_df))
```

```python
result.to_csv('../data/fillNA_london_aq.csv', index=False)
```
